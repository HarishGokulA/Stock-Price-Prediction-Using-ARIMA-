STOCK PREDICTION MODEL BY USING ARIMA MODEL

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
!pip install statsmodels
import yfinance as yf
!pip install statsmodels matplotlib
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from time import time
import datetime
import warnings

warnings.filterwarnings('ignore')

tickersymbol = "TVSMOTOR.NS"
data = yf.Ticker(tickersymbol)
data
end_date = datetime.date.today()
start_date = end_date - datetime.timedelta(days=1825)
tomorrow = end_date + datetime.timedelta(days=1)
prices = data.history(start=start_date, end=end_date).Close
prices
returns = prices.pct_change().dropna()
returns
plt.figure(figsize=(10,4))
plt.plot(prices)
plt.ylabel("Closing Prices")
plt.xlabel("Date")
plt.figure(figsize=(10,4))
plt.plot(returns)
plt.ylabel("returns")
plt.xlabel("Date")

plot_acf(returns)
plt.show()

plot_pacf(returns , method='ywm')
plt.show()
from statsmodels.tsa.arima.model import ARIMA

model = ARIMA(prices,order=(7,0,6))
fitted = model.fit()
print(fitted.summary())

next_day_price = fitted.forecast (100,alpha=0.05)
print(next_day_price)

predicted_price = next_day_price.values[0]
print(f"the predicted price for {tomorrow} tomorrow:", predicted_price)












```
